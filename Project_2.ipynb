#### Write code to guess random number generated by 

```python
import random

class RandomNumber:
    def __init__(self):
        self.number = random.randint(0, 100)
        self.guess = None
        self.count = 1

    def set_guess(self):
        self.guess = int(input("Enter the number (within 1000): "))

    def guess_the_number(self):
        while self.guess != self.number:
            if self.guess > self.number:
                print("You guessed a greater number.\nTry a smaller one.\n")
            else:
                print("You guessed a smaller number.\nTry a larger one.\n")
            
            self.count += 1
            self.set_guess()
        
        print(f"Congrat's!You guessed the number in {self.count}th attempt.")
        
    def feedback(self):
        if self.count == 1:
            print("Brilliant mind!")
        elif 1 < self.count <= 5:
            print("You have an excellent mind.")
        elif 5 < self.count <= 12:
            print("You have a very good mind.")
        else:
            print("You have a good mind.")

def main():
    rn = RandomNumber()
    rn.set_guess()
    rn.guess_the_number()
    rn.feedback()

if __name__ == "__main__":
    main()
```
